A better way to maximize the "cloud function" `f(x)` which is piecewise linear concave and has the special property that taking the maximum within each piece yields a set of points that define a quasiconcave function when joined together. 

Assuming an oracle that computes the maximum within each segment, this would be $\mathcal{O}(n)$, because there are $n$ segments. But the quasiconcavity property means we can use binary search instead, reducing the complexity to $\mathcal{O}(\log n)$.

```julia
# Small example
# n_segments = 10
# splits = [0, 0.08, 0.18, 0.2, 0.36, 0.47, 0.61, 0.68, 0.81, 0.9, 1]
# offsets = 0.001*[37, 47, 55, 61, 65, 67, 67, 65, 61, 55, 47]
# noise = [0.021, 0.023, 0.053, 0.004, 0.073, 0.007, 0.073, 0.06, 0.05, 0.052]


n_segments = 50
splits = vcat(0, sort(rand(n_segments-1)), 1)
offsets = collect(- range(0, 1, length=n_segments+1) .^2 + range(0, 1, length=n_segments+1))
noise = rand(n_segments) ./ n_segments

function f(x)
    idx = searchsortedfirst(splits[2:end-1], x)
    return offsets[idx] .- (x .- 0.5(splits[idx+1] + splits[idx]) - noise[idx])^2
end

plot(f, xlim=(0,1))
```

```julia
"""
    p(h)

When `h` is an index of which piece you are in, returns the optimal `x` and `f` value associated with that segment.
"""
function p(h)
    @assert h in 1:(length(splits)-1)
    
    x̄ = max(splits[h], 
            min(splits[h+1],
                0.5*(splits[h+1] + splits[h]) + noise[h]))
    return x̄, offsets[h] .- (x̄ .- 0.5(splits[h+1] + splits[h]) - noise[h])^2
end
```

The maximum is the *first* entry `h` of `L` such that `L[h+1] - L[h] ≤ 0`. We want to minimize calls to `L` in finding this.

```julia
"""
    binary_search(p, n)

Binary search for the global optimum in `p` when `p` indexes a function having `n` segments.
"""
function binary_search(p::Function, n)
    evaled_x = zeros(n+1)
    evaled_f = fill(Inf, n+1)
    evaled_f[n+1] = -Inf
    
    lb, ub = 1, n
    h = 0
    
    i= 0
    while ub - lb ≥ 2 && i < 25
        i += 1
        h = ub + (lb - ub) ÷ 2
        @show lb, ub, h
        if evaled_f[h]==Inf
            evaled_x[h], evaled_f[h] = p(h)
        end
        
        if evaled_f[h+1]==Inf
            evaled_x[h+1], evaled_f[h+1] = p(h+1)
        end
        
        if evaled_f[h+1] - evaled_f[h] ≤ 0
            ub = h
        else
            lb = h
        end
    end
    
    h = ub + (lb - ub) ÷ 2
    @show evaled_f
    return h, evaled_x[h], evaled_f[h]
end
```

```julia
h_star, x_star,  f_star = binary_search(p, n_segments)

pl = plot(xlim=(0,1), legend=:bottomright)

plot!(pl, f, x = 0:0.01:1, label="cloud function", c=:dodgerblue)
vline!(splits, ls=:dash, lw=0.4, label="segment limits", c=:olivedrab)
plot!(pl, p.(1:n_segments), ls=:dot, marker=:square, ms=3, msw=0, c=:darkorange, label="max on each segment")
scatter!(pl, [x_star], [f_star], marker=:star, ms=7, msw=0, c=:magenta, label="global max")

return pl
```

In the schema I am using to generate cloud functions, the function that appears below tends to be strictly concave for a large number of segments. However, it only has to be quasiconcave to use binary search.

```julia
pl = plot(legend=:bottomright)
plot!(pl, collect(1:n_segments), [p(h)[2] for h in 1:n_segments],
      ls=:dot, marker=:square, ms=3, msw=0, c=:darkorange, label="max on each segment")

scatter!(pl, [h_star], [f_star], marker=:star, ms=7, msw=0, c=:magenta, label="global max")
```

```julia
tabl = plot(layout=2, size=(1000, 400))

let pl = tabl[1]
    plot!(pl, xlim=(0,1), legend=:bottomright)
    plot!(pl, f, x = 0:0.01:1, label="cloud function f(x)", c=:dodgerblue)
    vline!(pl, splits, ls=:dash, lw=0.4, label="segment limits", c=:olivedrab)
    plot!(pl, p.(1:n_segments), ls=:dot, marker=:square, ms=3, msw=0, c=:darkorange, label="x⋆(h), f(x⋆(h))")
    scatter!(pl, [x_star], [f_star], marker=:star, ms=7, msw=0, c=:magenta, label="global max")
end

let pl = tabl[2]
    plot!(pl, legend=:bottomright)
    plot!(pl, collect(1:n_segments), [p(h)[2] for h in 1:n_segments],
          ls=:dot, marker=:square, ms=3, msw=0, c=:darkorange, label="h, f(x⋆(h))")
    scatter!(pl, [h_star], [f_star], marker=:star, ms=7, msw=0, c=:magenta, label="global max")
end

savefig(tabl, "cloud-surfing.pdf")
savefig(tabl, "cloud-surfing.png")

tabl
```
